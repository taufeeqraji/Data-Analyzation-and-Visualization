```python
# Import all data analytical modules and plotting modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import pie, axis, show
%matplotlib inline 
# for the graphs to be displayed as output
```

```python
# Open the csv file and redaing the data
# This is a Supermarket Sales Data csv file

df = pd.read_csv('supermarket_sales.csv')

# Displaying first 10 records
df.head(10)
```

```python
# This is a bar chart depicting the sales per branch
# bar plot using seaborn module

TotalPerBranch=df.groupby('Branch')['Total'].sum()
print(type(TotalPerBranch))
print(TotalPerBranch)
sns.barplot(x=TotalPerBranch.index , y=TotalPerBranch.values)
plt.ylabel('Total Sales')
plt.title('Total Sales per branch')
plt.xticks(rotation=45)
# displaying chart
plt.show()

# We can see that Branch C has the highest sale compared to branch A and B
# We can see sales in branch B and A are almost the same
# Maximum Sales value is over 100000
```

```python
#This is a multiple bar chart showing sales per branch by different payment modes
#countplot using seaborn module

sns.countplot(x='Branch',hue='Payment', data=df)
plt.xlabel('Branch Name')
plt.ylabel('Count')
plt.title('Sales by branch')
# displaying chart
plt.show()

# We can see that Branch A has the highest Ewallet payment
# Branch C has the highest Cash Payment
# All 3 modes are used in all 3 branches
```

```python
# Double bar graph showing sales per product type comparing with gender
# double bar graph using seaborn module

plt.figure(figsize=(15,5))
sns.countplot(data=df,x="Product line",hue="Gender",saturation=True,palette="Set2");

# We can see that Fashion accessories are the most purchased by Females
# and health and beauty are least purchased by females
# Health and beauty are most purchased by males
# and sports and travel are least purchased by males
```

```python
print(df.groupby('Product line')['City'].value_counts())

# Triple bar chart depicting the sales per product in all 3 cities

plt.figure(figsize=(15,5))
sns.countplot(data=df,x="Product line",hue="City",saturation=True,palette="Set1");

# We can see that Yangon City has the highest sale for Home and Lifestyle products
# Naypyitaw city has the highest sale for Food and beverages
# Mandalay city has the highest sale for Sports and travel
```

```python
print(df.groupby('City')['Product line'].value_counts())

# Bar chart comparing the sales of different products in each city

plt.figure(figsize=(15,5))
sns.countplot(data=df,x="City",hue="Product line",saturation=True,palette="Set2");

# We can infer that in Yangon city, home and lifestyle are most purchased
# In Naypyitaw city, food and beverages are most purchased
# In Mandalay city, Fashion accessories and Sports and travel are equally most purchased
```

```python
# This is a scatterplot depicting the total sales per city with gender

sns.scatterplot(x="City",y="Total",data=df,hue='Gender')
# displaying chart
plt.show()

# As we can see more orange dots in Naypyitaw City, we can infer that Males have more sales in Naypyitaw city 
# compared with females
```

```python
sums = df.groupby(df["Gender"])["Total"].sum()
tax = df.groupby(df["Gender"])["Tax 5%"].sum()
print(sums)
print(tax)
df3 = pd.DataFrame({
    'Gender': ['Male','Female'],
    'Total':  sums-tax,
    'Tax':  tax
})
print(df3)
sns.set(style='white')

# This is a Stacked plot for the total sale and tax by Male and Female

ax = df3.set_index('Gender').plot(kind='bar', stacked=True, color=['steelblue', 'red'])

# As males have purchased more overall, tax paid by males is higher than females
```

```python
# This is a pie chart showing the sale in each branch

sums = df.groupby(df["Branch"])["Total"].sum()

palette_color = sns.color_palette('bright')
  
explode = [0, 0, 0.1]
plt.pie(sums, labels=sums.index, colors=palette_color, explode=explode, autopct='%.0f%%')
  
# displaying chart
plt.show()

# We can see that branch C has the highest sale
# Branch A and B have equal sales
```

```python
# This is a Bar chart depicting the soverall ale of each Product type on all cities

plt.subplots(figsize=(12,8))
sns.countplot(x= 'Product line', data = df)

# We can infer that Fashion accessories has the highest sale
# and health and beauty has the least sale
```

```python
# This is a pie chart for the sale of different products

gross_sum = df.groupby(df["Product line"])["gross income"].sum()
# define Seaborn color palette to use
palette_color = sns.color_palette('colorblind')
  
# plotting data on chart
plt.pie(gross_sum, labels=gross_sum.index, colors=palette_color, autopct='%.0f%%')
  
# displaying chart
plt.show()

# We can infer that health and beauty has the least sale
# and other products have equally more sale
```

```python
best_prod = df[df.Rating >=8]
best_prod.groupby('Product line')['Rating'].count().sort_values(ascending=False)
plt.subplots(figsize=(12,8))
sns.countplot(x= 'Product line', data = best_prod)
```

```python
# This is a histogram for the freqency of the Quantity of products purchased

df['Quantity'].plot(kind='hist')

# We can see that 9-10 number of products purchased has been the highest frequency
```

```python
# This is a histogram showing the frequency of the total amount purchased

df['Total'].plot(kind='hist')

# We can see that 0-100 has been the frequently purchased amount
```

```python
# This is a distplot for the total sales

sns.distplot(df['Total'])
```

```python
# This is a distplot for the Rating

sns.distplot(df['Rating'])
```

```python
# This is a jointplot depicting the total amount and the quantity purchased

sns.jointplot( df['Total'],df['Quantity'])
```

```python
# We take out the year alone from the date to compare the sales in two different years

df['Year']=df['Date'].apply(lambda x: (str(x)[-4:len(x):1]))
print(df['Year'])

# Bar chart depicting the sale in two different years

sns.barplot(x=df['Year'] , y=df['Total'])
plt.ylabel('Total Sales')
plt.title('Total Sales in two different years')
# displaying chart
plt.show()

# We can see that there has been more sale in the year 2019 than in 2020
```
